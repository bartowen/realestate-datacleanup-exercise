#### Ejercicio 18. Normaliza la variable de precios para cada población y traza los 4 histogramas en el mismo gráfico (debes usar el subconjunto obtenido en la pregunta 14) (★★★)

Para el método de normalización, puedes usar el que consideres adecuado, no hay una única respuesta correcta para esta pregunta. Imprime el gráfico y escribe en la celda de Markdown un breve análisis sobre el gráfico.

Pista: Puedes ayudarte revisando la demostración multihist de Matplotlib.

```python
# TODO

from sklearn.preprocessing import MinMaxScaler
import matplotlib.pyplot as plt

# Copia df
df_normalizado = df_sur_madrid.copy()

scaler = MinMaxScaler()

# Normalizar los precios por población
df_normalizado['precio_normalizado'] = df_sur_madrid.groupby('level5')['price'].transform(
    lambda x: scaler.fit_transform(x.values.reshape(-1, 1)).flatten()
)

# Histograma
plt.figure(figsize=(12, 7))

for poblacion in df_normalizado['level5'].unique():
    subset = df_normalizado[df_normalizado['level5'] == poblacion]
    plt.hist(subset['precio_normalizado'], bins=20, alpha=0.5, label=poblacion)

plt.title('Histogramas de Precios Normalizados por Población (Cinturón Sur de Madrid)')
plt.xlabel('Precio Normalizado')
plt.ylabel('Frecuencia')
plt.legend()
plt.show()


```

Las cuatro poblaciones tienen una distribución sesgada a la derecha, lo que indica que la mayoría de las viviendas tienen precios más bajos (cerca de 0), mientras que hay pocas viviendas con precios más altos.

Las viviendas más caras son excepciones (outliers)

Las poblaciones como Alcorcón y Leganés muestran preciosmás bajos (cerca de 0.1 y 0.2).

Getafe, muestra una distribución con una mayor frecuencia en valores más cercanos a 0.4. Esto podría indicar que los precios son más elevados en esta región y comparados con las otras regiones.

#### Ejercicio 20. Realiza el mismo gráfico para 4 poblaciones diferentes (columna level5) y colócalos en el mismo gráfico. Debes usar el subconjunto obtenido en la pregunta 14 (★★☆)
